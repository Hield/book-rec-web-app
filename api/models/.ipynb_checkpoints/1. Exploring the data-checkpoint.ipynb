```python
m = {}
def count_string_combinations(title):
    for i in range(0, len(title) - 2):
        for j in range(i + 3, len(title) + 1):
            if title[i:j] not in m:
                m[title[i:j]] = []
            if title not in m[title[i:j]]:
                m[title[i:j]].append(title)
```

```python
books.title.apply(lambda x: count_string_combinations(x.lower()))
len(m)
```

```python
books.original_title.apply(lambda x: count_string_combinations(str(x).lower()))
len(m)
```

```python
class Book():
    def __init__(self, id, goodreads_id, authors, title, original_title,
                 average_rating, ratings_count, ratings_1, ratings_2,
                 ratings_3, ratings_4, ratings_5):
        self.id = id
        self.goodreads_id = goodreads_id
        self.authors_str = authors
        self.authors = authors.split(', ')
        self.title = title
        self.original_title = original_title
        self.average_rating = average_rating
        self.ratings_count = ratings_count
        self.ratings_1 = ratings_1
        self.ratings_2 = ratings_2
        self.ratings_3 = ratings_3
        self.ratings_4 = ratings_4
        self.ratings_5 = ratings_5

    def __repr__(self):
        return self.title + ' by ' + self.authors_str
    
    def __eq__(self, other):
        if isinstance(other, Book):
            return self.id == other.id
        return False

class BookDatabase():
    id_dict = {}
    search_dict = {}

    @classmethod
    def initialize_dicts(cls):
        books = pd.read_csv('./data/books.csv')
        books.apply(cls._add_book_from_pd, axis=1)
        books.apply(cls._add_book_for_searching, axis=1)

    @classmethod
    def _add_book_from_pd(cls, book):
        cls.id_dict[book.book_id] = Book(book.book_id, 
            book.goodreads_book_id, book.authors, book.title, book.original_title, book.average_rating,
            book.ratings_count, book.ratings_1, book.ratings_2, book.ratings_3,
            book.ratings_4, book.ratings_5)

    @classmethod
    def _add_book_for_searching(cls, book):
        title = book.title.lower()
        title_length = len(title)
        book_id = book.book_id
        for i in range(0, title_length - 2):
            for j in range(i + 3, title_length + 1):
                search_string = title[i:j]
                if search_string not in cls.search_dict:
                    cls.search_dict[search_string] = []
                if cls.id_dict[book_id] not in cls.search_dict:
                    cls.search_dict[search_string].append(cls.id_dict[book_id])

    @classmethod
    def search(cls, search_string):
        search_string = search_string.lower()
        if search_string not in cls.search_dict:
            return 'Not found'
        return cls.search_dict[search_string]
```

```python
%%time
BookDatabase.initialize_dicts()
```

```python
BookDatabase.search('hunger')
```
